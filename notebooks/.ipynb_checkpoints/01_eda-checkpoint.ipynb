```python
# 01_EDA.ipynb

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set(font_scale=1.1)

# -----------------------
# 1. Paths & Column Names
# -----------------------
DATA_DIR = "../data/raw"   # <--- important

train_path = os.path.join(DATA_DIR, "train_FD001.txt")
test_path  = os.path.join(DATA_DIR, "test_FD001.txt")
rul_path   = os.path.join(DATA_DIR, "RUL_FD001.txt")

col_names = ["unit", "time", "op_set_1", "op_set_2", "op_set_3"] + \
            [f"sensor_{i}" for i in range(1, 22)]

# -----------------------
# 2. Load Raw Data
# -----------------------
train_raw = pd.read_csv(train_path, sep=r"\s+", header=None,
                        names=col_names, engine="python")
test_raw = pd.read_csv(test_path, sep=r"\s+", header=None,
                       names=col_names, engine="python")
rul_raw = pd.read_csv(rul_path, sep=r"\s+", header=None, names=["RUL"])

print("Train shape:", train_raw.shape)
print("Test shape :", test_raw.shape)
print("RUL shape  :", rul_raw.shape)

train_raw.head()
```

```python
# 3. Basic data checks

print("\nTrain nulls:")
print(train_raw.isna().sum())

print("\nSensor value ranges:")
print(train_raw.describe().T[['mean', 'std', 'min', 'max']].head(10))
```

```python
# 4. Distribution of operating settings

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, col in enumerate(["op_set_1", "op_set_2", "op_set_3"]):
    sns.histplot(train_raw[col], kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# 5. Example sensor behavior over time for a single engine

example_unit = 1
cols_to_plot = ["sensor_2", "sensor_3", "sensor_4", "sensor_11"]

df_u = train_raw[train_raw["unit"] == example_unit]

plt.figure(figsize=(12, 6))
for col in cols_to_plot:
    plt.plot(df_u["time"], df_u[col], label=col)
plt.xlabel("Time (cycles)")
plt.ylabel("Sensor value")
plt.title(f"Sensor traces for unit {example_unit}")
plt.legend()
plt.show()
```
